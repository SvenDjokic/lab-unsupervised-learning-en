```python
# Pareto Principle Analysis

spending_columns = ["Fresh", "Milk", "Grocery", "Frozen", "Detergents_Paper", "Delicassen"]

customers_df["Total_Spending"] = customers_df[spending_columns].sum(axis=1)

customers_df_sorted = customers_df.sort_values('Total_Spending', ascending=False)

print(customers_df_sorted.head())

plt.figure(figsize=(10, 6))
sns.lineplot(data=customers_df_sorted['Total_Spending'], label="Total Spending")
plt.axhline(80, color='r', linestyle='--', label="80% Line")
plt.legend()
plt.title("Pareto Principle Check")
plt.show()
```

```python
# Calculate the cumulative spending
customers_df_sorted["Cumulative Spending"] = customers_df_sorted["Total_Spending"].cumsum()

# Calculate the cumulative percentage
total_spending = customers_df["Total_Spending"].sum()
customers_df_sorted["Cumulative Percentage"] = (customers_df_sorted["Cumulative Spending"] / total_spending) * 100

plt.figure(figsize=(10, 6))
sns.lineplot(data=customers_df_sorted["Cumulative Percentage"], label="Cumulative Spending")
plt.axhline(80, color='r', linestyle='--', label="80% Line")  # Mark 80%
plt.axvline(len(customers_df_sorted) * 0.2, color='g', linestyle='--', label="20% Line")  # Mark 20% customers
plt.title("Pareto Principle Check (Cumulative)")
plt.xlabel("Customers (Sorted by Spending)")
plt.ylabel("Cumulative Percentage of Spending")
plt.legend()
plt.show()
```

```python
# Total spending (replace 'spending_column' with your column name)
total_spending = customers_df['spending_columns'].sum()
customers_df['cumulative_percentage'] = (customers_df['spending_columns'].cumsum() / total_spending) * 100

# Sort and check Pareto principle
customers_df_sorted = customers_df.sort_values('spending_columns', ascending=False)
plt.figure(figsize=(10, 6))
sns.lineplot(data=customers_df_sorted['cumulative_percentage'], label="Cumulative % of Spending")
plt.axhline(80, color='r', linestyle='--', label="80% Line")
plt.legend()
plt.title("Pareto Principle Check")
plt.show()
```

```python
top_20_percent = int(len(customers_df_sorted) * 0.2)
spending_by_top_20 = customers_df_sorted["Total_Spending"].iloc[:top_20_percent].sum()
percentage_contributed = (spending_by_top_20 / total_spending) * 100
print(f"Top 20% of customers contribute {percentage_contributed:.2f}% of total spending.")
```

# Challenge 3 - Data Preprocessing

One problem with the dataset is the value ranges are remarkably different across various categories (e.g. `Fresh` and `Grocery` compared to `Detergents_Paper` and `Delicassen`). If you made this observation in the first challenge, you've done a great job! This means you not only completed the bonus questions in the previous Supervised Learning lab but also researched deep into [*feature scaling*](https://en.wikipedia.org/wiki/Feature_scaling). Keep on the good work!

Diverse value ranges in different features could cause issues in our clustering. The way to reduce the problem is through feature scaling. We'll use this technique again with this dataset.

#### We will use the `StandardScaler` from `sklearn.preprocessing` and scale our data. Read more about `StandardScaler` [here](https://scikit-learn.org/stable/modules/generated/sklearn.preprocessing.StandardScaler.html#sklearn.preprocessing.StandardScaler).

*After scaling your data, assign the transformed data to a new variable `customers_scale`.*

```python
# Import library
from sklearn.preprocessing import OneHotEncoder

# Perform one-hot encoding on "Channel" and "Region"
customers_scale = pd.get_dummies(customers_df_cleaned, columns=["Channel", "Region"], drop_first=True)

# Standardize spending columns
from sklearn.preprocessing import StandardScaler

spending_columns = ["Fresh", "Milk", "Grocery", "Frozen", "Delicassen"]
scaler = StandardScaler()
customers_scale[spending_columns] = scaler.fit_transform(customers_scale[spending_columns])

# Check the final DataFrame
print(customers_scale.head())
```

# Challenge 4 - Data Clustering with K-Means

Now let's cluster the data with K-Means first. Initiate the K-Means model, then fit your scaled data. In the data returned from the `.fit` method, there is an attribute called `labels_` which is the cluster number assigned to each data record. What you can do is to assign these labels back to `customers` in a new column called `customers['labels']`. Then you'll see the cluster results of the original data.

```python
print(customers_scale.shape)
print(customers_df.shape)
```

```python
from sklearn.cluster import KMeans


number_clusters = 3  
kmeans = KMeans(n_clusters=number_clusters, random_state=42)
kmeans.fit(customers_scale)

# Add cluster labels to the DataFrame
customers_scale['labels'] = kmeans.labels_

# Check the resulting clusters
print(customers_scale.head())
```

# Challenge 5 - Data Clustering with DBSCAN

Now let's cluster the data using DBSCAN. Use `DBSCAN(eps=0.5)` to initiate the model, then fit your scaled data. In the data returned from the `.fit` method, assign the `labels_` back to `customers['labels_DBSCAN']`. Now your original data have two labels, one from K-Means and the other from DBSCAN.

```python
from sklearn.cluster import DBSCAN

# Initialize DBSCAN with eps=0.5
dbscan = DBSCAN(eps=0.5)

# Fit the model to the scaled data
dbscan.fit(customers_scale)

# Assign the labels back to the DataFrame
customers_scale['labels_DBSCAN'] = dbscan.labels_

```
